```python
for x in data.columns:
    if data[x].dtype=='object' or data[x].dtype=='bool':
        data[x].fillna(data[x].mode()[0],inplace=True)
    elif data[x].dtype=='int64' or data[x].dtype=='float64':
        data[x].fillna(round(data[x].mean()),inplace=True)
```

```python
# Dropping the "CUST_ID" columns
data=data.drop('CUST_ID',axis=1)
```

```python
data.isnull().sum()
```

```python
for i in data.columns:     # For Boxplot --> use this method because it will show each boxplot indivisually.
    data.boxplot(column=i)
    plt.show()
```

```python
# Identify outliers / Detecting the Outliers.
q1 = (data["BALANCE"].quantile(0.25)) #first quartile value
q3 = (data["BALANCE"].quantile(0.75)) # third quartile value
iqr = q3-q1                #Interquartile range
low  = q1-1.5*iqr          #acceptable range
high = q3+1.5*iqr          #acceptable range
```

```python
data_include = data["BALANCE"].loc[(data["BALANCE"] >= low) & (data["BALANCE"] <= high)] # meeting the acceptable range
data_exclude = data["BALANCE"].loc[(data["BALANCE"] < low) | (data["BALANCE"] > high)] #not meeting the acceptable range
```

```python
data.boxplot(column='BALANCE') 
plt.show()
```

```python
corr_df=data.corr(method="pearson") 
print(corr_df)
 
sns.heatmap(corr_df,vmax=1.0,vmin=-1.0,annot=True) 
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 0.95)
X_train_new = pca.fit_transform(data)
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```

```python
from sklearn.cluster import KMeans
 
wsse = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i,  random_state = 10)
    kmeans.fit(data)
    wsse.append(kmeans.inertia_)
plt.plot(range(1, 11), wsse)
# plt.scatter(range(1, 11),wsse)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WSSE')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5,  random_state = 10)
Y_pred = kmeans.fit_predict(data)
```

```python
kmeans.inertia_  # Inertia --> WSSE at each K
```

```python
# How many iterations performed to reach the Zero Conversion State
kmeans.n_iter_
```
